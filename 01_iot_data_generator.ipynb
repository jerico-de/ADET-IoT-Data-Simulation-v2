This script generates simulated IoT sensor data for a Smart Logistics Tracking system.
It produces 100 records across 3 sensor types: Temperature, Humidity, and GPS.

---
### Step 1 – Import Libraries

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

print('✅ Libraries imported successfully!')
```

---
### Step 2 – Configure Parameters

```python
# ── CONFIGURATION ────────────────────────────────────────────────────
NUM_RECORDS  = 100                        # Total IoT records to generate
START_DATE   = datetime(2026, 1, 1, 0, 0, 0)
END_DATE     = datetime(2026, 1, 10, 23, 59, 59)   # 10-day window
STATUSES     = ['In Transit', 'Delayed', 'Delivered']
STATUS_PROBS = [0.4, 0.4, 0.2]           # Probability weights per status

# Sensor type distribution across 100 records
# 34 Temperature + 33 Humidity + 33 GPS = 100
sensor_pool = ['Temperature'] * 34 + ['Humidity'] * 33 + ['GPS'] * 33
np.random.shuffle(sensor_pool)

# Generate 100 random, sorted timestamps within the window
total_seconds_range = int((END_DATE - START_DATE).total_seconds())
random_offsets = np.random.randint(0, total_seconds_range, size=NUM_RECORDS)
random_offsets.sort()
RANDOM_TIMESTAMPS = [START_DATE + timedelta(seconds=int(off)) for off in random_offsets]
# ─────────────────────────────────────────────────────────────────────

print(f'✅ Configuration set:')
print(f'   Total Records : {NUM_RECORDS}')
print(f'   Sensor Types  : Temperature (34), Humidity (33), GPS (33)')
print(f'   Date Range    : {START_DATE.date()} to {END_DATE.date()}')
```

---
### Step 3 – Generate IoT Records

```python
# Smart Logistics IoT Data Generator
data = []

for i in range(NUM_RECORDS):
    timestamp = RANDOM_TIMESTAMPS[i]
    pkg_id    = np.random.randint(1000, 1100)
    rfid      = f'RFID_{i}'                        # device_id
    lat       = round(np.random.uniform(10.0, 15.0), 6)  
    lon       = round(np.random.uniform(120.0, 125.0), 6) 
    status    = np.random.choice(STATUSES, p=STATUS_PROBS)
    sensor    = sensor_pool[i]                    # data_type

    # Generate data_value based on sensor type
    if sensor == 'Temperature':
        raw_val    = round(np.random.uniform(2.0, 8.0), 2)  # Cold chain range (°C)
        data_value = f'{raw_val}C'
        num_val    = raw_val
    elif sensor == 'Humidity':
        raw_val    = round(np.random.uniform(40.0, 90.0), 2)  # Humidity (%)
        data_value = f'{raw_val}%'
        num_val    = raw_val
    else:  # GPS
        raw_val    = round(np.sqrt(lat**2 + lon**2), 6)  # Distance from origin
        data_value = f'{lat},{lon}'
        num_val    = raw_val

    record = {
        'timestamp':     timestamp.strftime('%Y-%m-%d %H:%M:%S'),
        'device_id':     rfid,
        'package_id':    pkg_id,
        'data_type':     sensor,
        'data_value':    data_value,
        'numeric_value': num_val,
        'latitude':      lat,
        'longitude':     lon,
        'status':        status
    }
    data.append(record)

# Convert to DataFrame
df = pd.DataFrame(data)

print(f'✅ Generated {len(df)} IoT records successfully!')
print(f'   Sensor breakdown: {df["data_type"].value_counts().to_dict()}')
```

```python
# Save as CSV and JSON
df.to_csv('logistics_iot_data.csv', index=False)
df.to_json('logistics_iot_data.json', orient='records', indent=2)

print('✅ Dataset saved successfully!')
print('   logistics_iot_data.csv  — for blockchain ledger')
print('   logistics_iot_data.json — backup format')
print(f'\nDataset shape: {df.shape[0]} rows x {df.shape[1]} columns')
print(f'Columns: {list(df.columns)}')
```
